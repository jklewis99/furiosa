```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import the dataset

```python
dataset = pd.read_csv("dbs/movies-from-2010s.csv")
dataset = dataset[['budget', 'revenue']]
dataset = dataset.loc[dataset['revenue'] > 0]
dataset = dataset.loc[dataset['budget'] > 0]
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
predictions = regressor.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title("Revenue vs Budget (training set)")
plt.xlabel("Budget (hundred millions)")
plt.ylabel("Revenue (Billions)")
plt.show()
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, predictions, color='blue')
plt.title("Revenue vs Budget (training set)")
plt.xlabel("Budget (hundred millions)")
plt.ylabel("Revenue (Billions)")
plt.show()
```
