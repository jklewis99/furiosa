## Find missing values in each dataframe

First, let's write a function that we can use to show our missing data

```python
def is_zero(series_object, columns):
    for i, value in enumerate(series_object):
        series_object[i] = value == 0
    return series_object

def display_zero_data(df):
    columns = df.columns
    percent_missing = 100 * df.copy().apply(lambda series: is_zero(series, columns), axis=1).sum() / len(df)
    zero_value_df = pd.DataFrame({'column': df.columns,
                                 'percent': percent_missing})
    zero_value_df.sort_values('percent', inplace=True)
    zero_value_df.reset_index(drop=True, inplace=True)
    zero_value_df = zero_value_df[zero_value_df['percent'] > 0]

    # Plotting
    fig = px.bar(
        zero_value_df, 
        x='percent', 
        y="column", 
        orientation='h', 
        title='Columns with Values that Equal 0', 
        height=max(len(zero_value_df) * 100, 200),
        width=600
    )
    fig.show()

def display_missing_data(df):
    percent_missing = 100 * df.isnull().sum() / len(df)
    missing_value_df = pd.DataFrame({'column': df.columns,
                                 'percent': percent_missing})
    missing_value_df.sort_values('percent', inplace=True)
    missing_value_df.reset_index(drop=True, inplace=True)
    missing_value_df = missing_value_df[missing_value_df['percent'] > 0]
    # Plotting
    fig = px.bar(
        missing_value_df, 
        x='percent', 
        y="column", 
        orientation='h', 
        title='Columns with Missing Values', 
        height=max(len(missing_value_df) * 100, 200),
        width=600
    )
    fig.show()
```

### Most prevalent genres

```python
from wordcloud import WordCloud, ImageColorGenerator
text = ""
for genre in movies_2010s.genres:
    # we want to generate a long String of all the genres for input to wordcloud. In the dataset, Sci-fi is formatted as such, so we will replace the - with an empty space so it is no interpreted as a space
    text += " ".join([g.replace("-", "") for g in genre.split("|")])
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap="magma", background_color="black", random_state=0).generate(text)
plt.figure(figsize=(15,10))
# Display the worcloud image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Most prevalent directors

```python
from collections import Counter
list_names = [name for name in directors.name]
word_count_dict = Counter(list_names)
wordcloud = WordCloud(colormap="magma", background_color="black", random_state=0).generate_from_frequencies(word_count_dict)
plt.figure(figsize=(15,10))
# Display the worcloud image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Is there a linear correlation between budget and revenue?

```python
dataset = pd.read_csv("dbs/movies_from_2010s.csv")[['budget', 'revenue']]
dataset = dataset.loc[dataset['budget'] > 0]
budget = dataset.iloc[:, 0].values
revenue = dataset.iloc[:, -1].values

plt.scatter(budget, revenue, color='red')
plt.title("Revenue vs Budget")
plt.xlabel("Budget (hundred millions)")
plt.ylabel("Revenue (Billions)")
plt.show()
```
